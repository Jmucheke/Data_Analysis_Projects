```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#  getting a paragraph
soup.find_all('p', class_="lead")
```

```python
# Extracting text from the paragraph
soup.find('p', class_="lead").text.strip()
```

```python
# Getting table headers
soup.find_all('th')
```

```python
soup.find_all('th')[0].text.strip()
```

```python
table = soup.find('table', class_='table')
table
```

```python
# Getting table headers
table_headers = table.find_all('th')
table_headers = [header.text.strip() for header in table_headers]
table_headers
```

```python
import pandas as pd

df = pd.DataFrame(columns=table_headers)
df
```

```python
# Getting table rows and table data
column_data = table.find_all('tr')[1:]
column_data
```

```python
# Extracting data from each row and populating the dataframe
for row in column_data:
    row_data = row.find_all('td')
    row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = row_data

print(df)
# saving the dataframe to a CSV file
df.to_csv('Hockey_teams.csv', index=False)
```

```python
# Getting all the data from multiple pages
all_data = pd.DataFrame(columns=table_headers)

for page in range(1, 24):
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    table = soup.find('table', class_='table')
    column_data = table.find_all('tr')[1:]
    
    for row in column_data:
        row_data = row.find_all('td')
        row_data = [data.text.strip() for data in row_data]
        length = len(all_data)
        all_data.loc[length] = row_data
        
all_data
```

```python
# Using while loop to get all the data from multiple pages
all_data = pd.DataFrame(columns=table_headers)

page = 1
while True:
    url = f"https://www.scrapethissite.com/pages/forms/?page_num={page}"
    response = requests.get(url)
    
    if response.status_code != 200:
        break
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    table = soup.find('table', class_='table')
    
    # break if no table is found
    if not table:
        break
    
    column_data = table.find_all('tr')[1:]
    
    if not column_data:
        break
    
    for row in column_data:
        row_data = row.find_all('td')
        row_data = [data.text.strip() for data in row_data]
        
        length = len(all_data)
        
        all_data.loc[length] = row_data
    
    page += 1
```

```python
# saving the dataframe to a CSV file
all_data.to_csv('Hockey_teams.csv', index=False)
```
