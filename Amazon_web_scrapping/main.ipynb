```python
# inport libraries
from bs4 import BeautifulSoup
import requests
from fake_useragent import UserAgent
import smtplib
import time
import datetime
```

```python
# connect to website
url = 'https://www.amazon.com/Apple-MacBook-Laptop-14%E2%80%91core-32%E2%80%91core/dp/B0DLHMYX53?th=1'

ua = UserAgent()

headers = {
    'User-Agent': ua.random,
    'Accept-Language': 'en-US,en;q=0.9',
}

response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.content, 'html.parser')

soup = BeautifulSoup(soup.prettify(), 'html.parser')

print(soup)
```

```python
title = soup.find(id='productTitle').text.strip()
title
```

```python
price = soup.find(class_='a-offscreen').text.strip()[1:].split('.')[0]
price
```

```python
Date = datetime.date.today()
```

```python
import csv

header = ['Title', 'Price', 'Date']
data = [title, price, Date]

with open('AmazonScrapperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
def send_mail():
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.ehlo()
    server.starttls()
    server.ehlo()
    server.login('email@gmail.com', 'password')
    
    subject = 'Price went down!'
    body = 'Hi, the price for the MacBook you want dropped below $2000. Check the amazon link https://www.amazon.com/Apple-MacBook-Laptop-14%E2%80%91core-32%E2%80%91core/dp/B0DLHMYX53?th=1'
    
    msg = f"Subject: {subject}\n\n{body}"
    
    server.sendmail('email@gmail.com', msg)
```

```python
def check_price():
    url = 'https://www.amazon.com/Apple-MacBook-Laptop-14%E2%80%91core-32%E2%80%91core/dp/B0DLHMYX53?th=1'

    ua = UserAgent()

    headers = {
        'User-Agent': ua.random,
        'Accept-Language': 'en-US,en;q=0.9',
    }

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.content, 'html.parser')

    soup = BeautifulSoup(soup.prettify(), 'html.parser')
    
    title = soup.find(id='productTitle').text.strip()
    
    price = soup.find(class_='a-offscreen').text.strip()[1:].split('.')[0]
    
    Date = datetime.date.today()
    
    data = [title, price, Date]
    
    with open('AmazonScrapperDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
        
    if (price < 2000):
        send_mail()
    
    
```

```python
while(True):
    check_price()
    time.sleep(86400)
```
