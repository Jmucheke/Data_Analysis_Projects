```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df1 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df1
```

```python
df1 = df1[sorted(df1.columns)]
```

```python
df1.columns
```

```python
df2 = df1.transpose()
plt.figure(figsize=(12, 8))
df2.plot()
plt.xticks(rotation=90)
plt.show()
```
